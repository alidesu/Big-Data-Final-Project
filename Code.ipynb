```python

# Convert 'Year' column to integer for consistency
df['Year'] = df['Year'].astype(int)

```

```python

# Count movies per year
yearly_counts = df['Year'].value_counts().sort_index().to_dict()
```

```python

# Prepare the output JSON format
yearly_trend = {str(year): count for year, count in yearly_counts.items()}
```

```python
import json 
```

```python

# Output the result
print(json.dumps({"yearlyTrend": yearly_trend}, indent=2))
```

```python
df = pd.read_csv('IMDb_Dataset_2.csv')
```

```python
df.head()
```

```python
movies_data = df[['Title', 'Director', 'Star Cast', 'Poster-src']]
```

```python
movies_json = movies_data.to_json(orient='records', lines=False)
```

```python
print(movies_json)
```

```python
with open('movies_data.json', 'w') as file:
    file.write(movies_json)
```

```python

# Extract the required columns
required_columns = ['Title', 'Director', 'Star Cast', 'Poster-src', 'IMDb Rating', 'Year', 'Genre']
movie_data = df[required_columns].to_dict('records')

# Transform the data to match the expected structure
movies = []
for movie in movie_data:
    movie_dict = {
        'movie_name': movie['Title'],
        'director': movie['Director'],
        'cast': movie['Star Cast'],
        'image_src': movie['Poster-src'],
        'rating': movie['IMDb Rating'],
        'year': movie['Year'],
        'genre': movie['Genre']
    }
    movies.append(movie_dict)

# Save the movies data to a JSON file
with open('movies_data.json', 'w') as f:
    json.dump(movies, f, indent=2)
```
